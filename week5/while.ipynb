```python
# write a program that convert temperature from celcius to farenhiet
# ask user to enter temperature in celsius then convert to farenhiet. then ask if user want to convert another temperature.
#if user choose 'yes' then do it again
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

while True:
    try:
        celsius = float(input("Enter temperature in Celsius: "))
        fahrenheit = celsius_to_fahrenheit(celsius)
        print(f"{celsius}째C is {fahrenheit}째F")
    except ValueError:
        print("Please enter a valid number.")
        continue
    
    again = input("Do you want to convert another temperature? (yes/no): ").lower()
    if again != 'yes':
        print("Goodbye!")
        break
```

```python
#a system is being tested to find severe bugs, each bugs is assigned to a fault-points.
#if total fault-points exceed 100, the system will stop.
#write a program to simulate that
import random

def test_system():
    total_fault_points = 0
    bug_count = 0
    
    while total_fault_points <= 100:
       
        fault_points = random.randint(1, 20)
        bug_count += 1
        total_fault_points += fault_points
        print(f"Bug {bug_count}: Fault points = {fault_points}, Total fault points = {total_fault_points}")
        
        if total_fault_points > 100:
            print("\nSystem stopped! Fault-point limit exceeded.")
            break

test_system()
```

```python
#system that reminds user to drink water after every 2h, starting from 8am.
#asking user amount of water they drink and showing the amount the user need to drink more to reach 2l
#if user drinks 2l, it will stop
import time

def water_reminder():
    total_water = 0
    target = 2000  # Target is 2000 ml (2 liters)
    current_time = 8  # Start at 8 AM
    
    while total_water < target:
        print(f"\nIt's {current_time}:00. Time to drink water!")
        try:
            water_drunk = float(input("Enter the amount of water you drank (in ml): "))
            total_water += water_drunk
            remaining_water = target - total_water
            
            if total_water >= target:
                print(f"\nCongratulations! You've reached your goal of 2 liters today.")
                break
            else:
                print(f"You've drunk {total_water} ml so far. You need {remaining_water} ml more to reach 2 liters.")
        except ValueError:
            print("Please enter a valid number.")

        #reminder after 2 hours
        current_time += 2
        if current_time >= 24:  # Reset the time if it goes past midnight
            current_time -= 24

        
        
        print("\n...2 hours later...\n")  # Placeholder for the 2-hour reminder

# Start the water reminder system
water_reminder()
```

```python
#police is looking for a criminal in a building with 5 floors, each floor have 4 rooms.
#they will search each room in each floor till they catch the criminal.
import random

def search_building():
    
    floors = 5
    rooms_per_floor = 6

    # 
    criminal_floor = random.randint(1, floors)
    criminal_room = random.randint(1, rooms_per_floor)
    
    print(f"The criminal is hiding on Floor {criminal_floor}, Room {criminal_room} (This is secret, police don't know yet)")

    
    for floor in range(1, floors + 1):
        for room in range(1, rooms_per_floor + 1):
            print(f"Searching Floor {floor}, Room {room}...")
            
            
            if floor == criminal_floor and room == criminal_room:
                print(f"Criminal found on Floor {floor}, Room {room}!")
                return


search_building()
```

```python
# rock paper scissors game 
# each game has maximum 5 rounds. computer will randomly choose one. who wins 3 rounds first will win the game.
# after each game ask user if they want to play again.
import random

choices = ["rock", "paper", "scissors"]

while True:
    user_score = 0
    computer_score = 0
    round_number = 1

    while round_number <= 5 and user_score < 3 and computer_score < 3:
        print(f"\n--- Round {round_number} ---")
        user_choice = input("Choose rock, paper, or scissors: ").lower()
        
        while user_choice not in choices:
            user_choice = input("Invalid choice. Please choose 'rock', 'paper', or 'scissors': ").lower()

        computer_choice = random.choice(choices)
        print(f"Computer chose: {computer_choice}")

        if user_choice == computer_choice:
            print("It's a tie!")
        elif (user_choice == "rock" and computer_choice == "scissors") or \
             (user_choice == "scissors" and computer_choice == "paper") or \
             (user_choice == "paper" and computer_choice == "rock"):
            print("You win this round!")
            user_score += 1
        else:
            print("Computer wins this round!")
            computer_score += 1

        print(f"Score - You: {user_score}, Computer: {computer_score}")
        round_number += 1

    if user_score == 3:
        print("\nCongratulations! You win the game!")
    elif computer_score == 3:
        print("\nSorry, the computer wins the game.")
    else:
        print("\nNo one reached 3 wins. It's a draw!")

    play_again = input("Do you want to play again? (yes/no): ").lower()
    if play_again != "yes":
        print("Thanks for playing! Goodbye!")
        break
```
